```python
import pandas as pd
import numpy as np
import pylab as plt
```

```python
d = pd.read_excel("表单合成.xlsx",header = 0,index_col = None)
d
```

```python
data = d.fillna(0)
data
```

```python
data = data.values
data
```

```python
D1 = []
for i in range(len(data)):
    if data[i,1] == '高钾' and data[i,2] == '无风化':
        D1.append(data[i])
D1
```

```python
for i in range(len(data)):
    if data[i,1] == '高钾' and data[i,2] == '风化':
        D1.append(data[i])
D1
```

```python
for i in range(len(data)):
    if data[i,1] == '铅钡' and data[i,2] == '无风化':
        D1.append(data[i])
D1
```

```python
for i in range(len(data)):
    if data[i,1] == '铅钡' and data[i,2] == '风化':
        D1.append(data[i])
D1
```

```python
data2 = np.array(D1)
data2
```
